```python
%matplotlib inline
import grale.lenses as lenses
import grale.plotutil as plotutil
import grale.feedback as feedback
import grale.images as images
from grale.constants import *
import numpy as np
import matplotlib.pyplot as plt

feedback.setDefaultFeedback("none")
plotutil.setDefaultAngularUnit(ANGLE_ARCSEC)

V = lambda x,y: np.array([x,y], dtype=np.double)
Dd, Ds, Dds = 1000*DIST_MPC, 1600*DIST_MPC, 1200*DIST_MPC
```

```python
lens = lenses.PlummerLens(Dd, { "mass": 1e14*MASS_SUN, "width": 15*ANGLE_ARCSEC})
src = images.CircularSource(V(5,0)*ANGLE_ARCSEC, 1.5*ANGLE_ARCSEC)
```

```python
plt.figure(figsize=(10,10))
plotutil.plotImagePlane(lens, sources=[src])

imgPos = [ 
    [ (-2.5,0), (-1.5, 0), (-2, 0.5), (-2, -0.5) ],
    [ (-20, 0), (-22, 0), (-20.5, 5), (-20.5,-5)],
    [ (28, 0), (27, 0), (27, 7), (27, -7)]
]
extImg = images.ImagesData(len(imgPos))
for i in range(len(imgPos)):
    img = imgPos[i]
    for pt in img:
        extImg.addPoint(i, V(*pt)*ANGLE_ARCSEC)
    
plotutil.plotImagesData(extImg)
#plt.gca().set_xlim([-6, 0])
#plt.gca().set_ylim([-3, 3])
```
